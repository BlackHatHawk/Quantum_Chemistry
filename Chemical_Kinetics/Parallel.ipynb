```python
import numpy as np
import matplotlib.pyplot as plt
```

A----->B    k1 = 0.1
A----->C    k2 = 0.12

```python
A = 1000
B = 0
C = 0
k1 = 0.1
k2 = 0.12
del_t = 0.001
T = 50
```

```python
seg = int(T/del_t)
Tt = np.linspace(0,T,seg)
At = np.zeros(seg)
Bt = np.zeros(seg)
Ct = np.zeros(seg)
At[0] = A
Bt[0] = B
Ct[0] = C
for i in range(1,seg):
    At[i] = At[i-1] - At[i-1] * k1*del_t - At[i-1] *k2*del_t
    Bt[i] = Bt[i-1] + At[i-1] * k1*del_t
    Ct[i] = Ct[i-1] + At[i-1] *k2*del_t
```

```python
plt.plot(Tt,At)
plt.plot(Tt,Bt)
plt.plot(Tt,Ct)
plt.xlabel('Time')
plt.ylabel('Concentration')
plt.legend(['A','B','C'])
```
